```python
import pandas as pd
```

```python
df = pd.read_csv("reviews_data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Removing the missing values in the row of Rating column

```python
df = df.dropna(subset=["Rating"])
```

```python
df.isnull().sum()
```

```python
df.head()
```

Removing the Reviewed word from the date column

```python
df["Date"] = df["Date"].str.replace("Reviewed ", "", regex=False)
```

```python
df.head()
```

```python
df.info()
```

Change the Date column datetype

```python
df["Date"] = pd.to_datetime(df["Date"], format="%b. %d, %Y", errors="coerce")
```

```python
df.head()
```

```python
df['Rating'] = df['Rating'].astype(int)
```

```python
df.head()
```

Create a Sentiment_Label column which stores the positive, negative and neutral value based on rating

```python
def rate_to_sentiment(rating):
    if rating >= 4:
        return "Positive"
    elif rating == 3:
        return "Neutral"
    else:
        return "Negative"

df["Sentiment_Label"] = df["Rating"].apply(rate_to_sentiment)
```

```python
df.head(15)
```

```python
df.to_csv("D: cleaned_reviews_data.csv", index=False)

print("Cleaned data saved successfully as 'cleaned_reviews_data.csv'")
```
